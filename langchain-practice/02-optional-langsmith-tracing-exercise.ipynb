# (선택 실습) LangSmith 트레이싱 연동

> 선택 사항 | `01-fact-checker.ipynb` 완료 후 진행

팩트체커 에이전트에 **LangSmith 트레이싱**을 연동하여 내부 동작을 시각적으로 추적합니다.  
이 실습은 추가 학습 Section 5(LangSmith 심화)를 위한 **맛보기**입니다.

### 이 실습에서 배우는 것
- LangSmith 프로젝트 생성 및 API 키 설정
- 기존 체인/에이전트에 코드 한 줄로 트레이싱 연동
- LangSmith UI에서 실행 흐름 추적
- 토큰 사용량 및 레이턴시 확인

> **참고**: LangSmith 무료 계정에서 월 5,000 트레이스까지 무료입니다.  
> 가입: https://smith.langchain.com

## 1. LangSmith 설정

### 1-1. API 키 발급

1. [smith.langchain.com](https://smith.langchain.com) 접속 및 가입
2. Settings → API Keys → Create API Key
3. 발급된 키를 `.env` 파일에 추가

```
# .env 파일에 추가
LANGCHAIN_TRACING_V2=true
LANGCHAIN_API_KEY=ls__your_key_here
LANGCHAIN_PROJECT=langchain-lecture-factchecker
```

## 2. 기존 팩트체커에 트레이싱 적용

LangSmith의 핵심 장점: **기존 코드를 전혀 수정하지 않아도 됩니다.**  
환경 변수만 설정하면 모든 LangChain/LangGraph 호출이 자동으로 추적됩니다.

## 3. 실행 이름 지정 — 추적 식별 용이하게

`langsmith_config`를 통해 각 실행에 이름을 붙이면 대시보드에서 찾기 쉬워집니다.

## 4. LangSmith UI에서 확인할 것들

[smith.langchain.com](https://smith.langchain.com) → 프로젝트 선택 → 방금 실행한 트레이스 클릭

### 확인 포인트

| 항목 | 위치 | 의미 |
|------|------|------|
| **실행 흐름** | 트레이스 트리 | Agent → Tool → Agent 순서 시각화 |
| **입/출력** | 각 노드 클릭 | 정확히 어떤 데이터가 오갔는지 |
| **토큰 사용량** | 상단 메타데이터 | 총 입력/출력 토큰 수 |
| **레이턴시** | 각 노드 우측 | 어느 단계에서 시간이 걸리는지 |
| **에러** | 빨간색 노드 | 실패한 Tool 호출 및 원인 |

### 디버깅에 활용하기

Agent가 예상과 다르게 동작할 때:
1. LangSmith에서 해당 트레이스 열기
2. Agent가 어떤 Tool을 선택했는지 확인
3. Tool에 전달된 입력과 반환된 결과 확인
4. LLM이 어떤 프롬프트를 받았는지 확인

> 이 수준의 가시성은 `print()` 디버깅으로는 불가능합니다.

## 5. 핵심 요약

| 개념 | 방법 |
|------|------|
| 트레이싱 활성화 | `.env`에 `LANGCHAIN_TRACING_V2=true` + `LANGCHAIN_API_KEY` 설정 |
| 기존 코드 수정 없음 | 환경 변수만으로 모든 LangChain 호출 자동 추적 |
| 커스텀 트레이스 | `@traceable` 데코레이터로 일반 함수도 포함 가능 |
| 프로젝트 분리 | `LANGCHAIN_PROJECT`로 실습/개발/운영 환경 분리 |
