### 4개 Phase를 통해 구축한 것

| Phase | 내용 | 핵심 개념 |
|-------|------|----------|
| 1 | 데이터 수집 노드 | StateGraph, Node, Edge |
| 2 | 수집→분석→리스크 플로우 | 조건부 엣지, 병렬 실행 |
| 3 | Supervisor 멀티에이전트 | 동적 라우팅, 에이전트 협업 |
| 4 | Human-in-the-loop + 리포트 | interrupt, checkpointer, streaming |

### 최종 아키텍처

```
사용자 요청
    |
    v
[Supervisor] ←───── 모든 에이전트 결과 수렴
    |
    |── [Researcher] : 웹 검색으로 데이터 수집
    |── [Quant]      : 정량 분석 (재무 지표)
    |── [Qual]       : 정성 분석 (뉴스 감성)
    |── [Risk]       : 리스크 평가
    |── [Reporter]   : 구조화된 리포트 생성
    |        ↑
    |   [Human 승인]
    |
    v
InvestmentReport (Pydantic)
```
