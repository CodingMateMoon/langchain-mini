## 4. LCEL 체인 구성

파이프 연산자(`|`)로 **Prompt → LLM → Parser** 체인을 만듭니다.

### 체인의 동작 원리

```
{"role": "...", "question": "..."}
        ↓
  [PromptTemplate]  → 변수를 채워 메시지 생성
        ↓
  [ChatUpstage]      → LLM에 메시지 전송, AIMessage 수신
        ↓
  [StrOutputParser]  → AIMessage에서 content 문자열 추출
        ↓
  "리스트 컴프리헨션은..."
```

## 7. 정리 및 핵심 요약

### 오늘 배운 것

| 개념 | 코드 | 역할 |
|------|------|------|
| ChatUpstage | `ChatUpstage(model="solar-pro")` | LLM 인스턴스 |
| ChatPromptTemplate | `ChatPromptTemplate.from_messages([...])` | 재사용 가능한 프롬프트 |
| StrOutputParser | `StrOutputParser()` | AIMessage → 문자열 변환 |
| LCEL 파이프 | `prompt \| llm \| parser` | 체인 구성 |
| invoke / stream / batch | `chain.invoke({...})` | 다양한 실행 방법 |
| RunnableParallel | `RunnableParallel(a=chain1, b=chain2)` | 병렬 실행 |
